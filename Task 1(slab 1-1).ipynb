```python
#IMPORTING LIBRARIES

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOAD THE CSV FILE
df = pd.read_csv("soft_drink_sales.csv")
```

```python
# FIRST FIVE ROWS OF DATAFRAME
df.head()
```

```python
# SUMMARY OF DATAFRAME
df.info()
```

```python
# ALL NON NUMERICAL COLUMNS STATISTICAL SUMMARY
df.describe(include="object")
```

```python
#   ALL NUMERICAL COLUMNS STATISTICAL SUMMARY
df.describe()
```

```python
# CHECK FOR ISNULLIN DATAFRAME
df.isnull().sum()
```

```python
#  AVERAGE REVENUE
average_revenue = df["Revenue"].mean()
average_revenue
```

```python
#  AVERAGE UNITS SOLD
average_units = df["Units Sold"].mean()
average_units
```

```python
#   BAR PLOT

company_sales = df.groupby("Company")["Revenue"].sum()

company_sales.plot(kind="bar")

plt.title("Revenue by Company")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.show()
```

#  IN THIS DATASET THE ABOVE GRAPH SHOWS COCAL-COLA COMPANY HAS MOST REVENUE THAN DR.PEPPER THEN PEPSI.

```python
#   BAR PLOT

company_units = df.groupby("Company")["Units Sold"].sum()

company_units.plot(kind="bar")

plt.title("Units sold  by Company")
plt.xlabel("Company")
plt.ylabel("Units sold")
plt.xticks(rotation=45)

plt.show()
```

# THE ABOVE GRAPH SHOWS COCA-COLA HAS SOLD MOST OF THEIR PRODUCTS .

```python
#   SCATTER PLOT

plt.scatter(df["Units Sold"], df["Revenue"])

plt.title("Units Sold vs Revenue")
plt.xlabel("Units Sold")
plt.ylabel("Revenue")

plt.show()
```

# THE ABOVE PLOT SHOWS  THE  RELATIONSHIP BETWEEN UNITS SOLD VS REVENUE BY COMPANY.
NOTE: AS FROM ABOVE GRAPH WE  COULDNT IDENTIFY MORE DETAILS SO BELOW I HAVE USED SEABORN FEATURE

```python
#SCATTER PLOT

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Units Sold", y="Revenue", hue="Category", palette="viridis")

plt.title("Units Sold vs Revenue by Category")
plt.xlabel("Units Sold")
plt.ylabel("Revenue")
plt.legend(title="Category")

plt.show()
```

# THE ABOVE GRAPH SHOWS THE RELATIONSHIP BETWEEN UNITS SOLD VS REVENUE BY CATEGORY OF COMPANY.

```python
#    HEATMAP

numerical_cols = ['Units Sold', 'Revenue', 'Cost of Goods Sold', 'Profit']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Sales Metrics")
plt.show()
```

#THE ABOVE GRAPH SHOWS THE RELATIONSHIP BETWEEN EVERY NUMERICAL COLUMN WITH ALL OTHER NUMERICAL COLUMN.

#BOXES WITH VALUE 1.00 OR POSITIVE VALUE SHOWS A STRONG POSITIVE REALTIONSHP MEANS IF ONE INCREASES OTHER ALSO INCREASES.
#Example: Every Column has a linear relationship with itself.

#BOXES WITH VALUE IN NEGATION SHOWS A NEGATIVE RELATIONSHIP MEANS IF ONE INCREASES OTHER DECREASES.

#BOXES WITH VALUE 0 SHOWS A ARBIRTATRY RELATIONSHIP NOT SURE.

# SUMMARY OF ABOVE:-
    The heatmap helps us quickly identify which sales metrics move together (positive correlation), which move in opposite directions (negative correlation), and which have little to no linear association.
