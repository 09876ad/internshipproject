```python
import numpy as np
```

```python
row1=int(input("enter no of rows:"))
column1=int(input("enter no of columns:"))
```

```python
l1 = [[0 for _ in range(column1 + 1)] for _ in range(row1 + 1)]

for i in range(1,row1+1):
  for j in range(1,column1+1):
    l1[i][j]=int(input(f"enter {i},{j} value:"))

matA=np.array(l1)
```

```python
row2=int(input("enter no of rows:"))
column2=int(input("enter no of columns:"))
```

```python
l2 = [[0 for _ in range(column2 + 1)] for _ in range(row2 + 1)]

for i in range(1,row2+1):
  for j in range(1,column2+1):
    l2[i][j]=int(input(f"enter {i},{j} value:"))

matB=np.array(l2)
```

```python
choice_int = int(input("enter 1:addition,2:subtraction,3:multiplication,4:transpose,5:determinant"))

if choice_int == 1:
  if matA.shape == matB.shape:
    result = np.add(matA, matB)
    print("Result of Addition:")
    print(result)
  else:
    print(f"Error: Cannot perform addition. Matrix A shape {matA.shape} and Matrix B shape {matB.shape} are not compatible for addition.")
elif choice_int == 2:
  if matA.shape == matB.shape:
    result = np.subtract(matA, matB)
    print("Result of Subtraction:")
    print(result)
  else:
    print(f"Error: Cannot perform subtraction. Matrix A shape {matA.shape} and Matrix B shape {matB.shape} are not compatible for subtraction.")
elif choice_int == 3:
  actual_matA = matA[1:, 1:]
  actual_matB = matB[1:, 1:]
  if actual_matA.shape[1] == actual_matB.shape[0]:
    result = np.dot(actual_matA, actual_matB)
    print("Result of Multiplication:")
    print(result)
  else:
    print(f"Error: Cannot perform multiplication. Columns of A ({actual_matA.shape[1]}) must match rows of B ({actual_matB.shape[0]}).")
elif choice_int == 4:
  transpose_choice = input("Enter 'A' to transpose Matrix A or 'B' to transpose Matrix B: ")
  if transpose_choice.upper() == 'A':
    actual_matA = matA[1:, 1:]
    result = np.transpose(actual_matA)
    print("Transpose of Matrix A:")
    print(result)
  elif transpose_choice.upper() == 'B':
    actual_matB = matB[1:, 1:]
    result = np.transpose(actual_matB)
    print("Transpose of Matrix B:")
    print(result)
  else:
    print("Invalid choice for transpose. Please enter 'A' or 'B'.")
elif choice_int == 5:
  det_choice = input("Enter 'A' to find determinant of Matrix A or 'B' to find determinant of Matrix B: ")
  if det_choice.upper() == 'A':
    actual_matA = matA[1:, 1:]
    if actual_matA.shape[0] == actual_matA.shape[1]:
      result = np.linalg.det(actual_matA)
      print("Determinant of Matrix A:")
      print(result)
    else:
      print(f"Error: Matrix A is not square ({actual_matA.shape}) and cannot have a determinant.")
  elif det_choice.upper() == 'B':
    actual_matB = matB[1:, 1:]
    if actual_matB.shape[0] == actual_matB.shape[1]:
      result = np.linalg.det(actual_matB)
      print("Determinant of Matrix B:")
      print(result)
    else:
      print(f"Error: Matrix B is not square ({actual_matB.shape}) and cannot have a determinant.")
  else:
    print("Invalid choice for determinant. Please enter 'A' or 'B'.")
else:
  print("Invalid choice. Please enter 1 for Addition, 2 for Subtraction, 3 for Multiplication, 4 for Transpose, or 5 for Determinant.")
```
